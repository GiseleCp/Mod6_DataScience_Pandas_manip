```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

# Manutenção

```python
dados = pd.json_normalize(dados['info_moveis'])
```

```python
# coletando o nome das colunas
colunas = list(dados.columns)
colunas
```

```python
# desagrupa os dados agrupados em lista
dados = dados.explode(colunas[3:]) # a partir do indice 3
dados.head()
```

```python
# como o indice ficou inconsistente, vamos resetar os dados da coluna de indice
dados.reset_index(inplace=True, drop=True) # inplace- substitui o indice atual drop-elimina coluna com indices antigos
dados.head()
```

```python
# info = retorna os tipos de dados na tabela
dados.info()
```

# Conversão números

```python
# para converter as colunas numericas int
import numpy as np
# altera o tipo do dados
dados['max_hospedes'].astype(np.int16)

# geralmente usa int64 para dados cientificos, para economizar memoria, podemos usar int16
```

```python
# a alteração não foi feita no dataframe
dados.info()
```

```python
# agora aplicamos a alteração no dataframe
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int16)
dados.info()
```

```python
# substituir várias colunas
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
dados[col_numericas] = dados[col_numericas].astype(np.int16)
dados.info()
```

```python
# para converter as colunas numericas float64
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float32)
dados.info()
```

```python
# para converter os valores, com o simbolo junto com valor e virgula, necessário tratar antes
dados['preco'] # o valor do campo é string

# uso de funcao lambda para converter
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float32)
dados.info()
```

```python
# para atualizar os valores multiplos de um dataframe, usamos applymap
col_valores = ['taxa_deposito', 'taxa_limpeza']
dados[col_valores] = dados[col_valores].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
dados[col_valores] = dados[col_valores].astype(np.float32)
dados.info()
```

# Conversão texto

```python
dados['descricao_local']
```

```python
# padronização em letras minusculas
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
# removendo - REGEX - caracteres especiais
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True) # apos ^ indica caracteres que não serão substituidos
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True) # apos ^ indic acaracteres que não serão substituidos
# (?<!\w) - na primeira parte, verifica se não há caracter antes do hifen - look behind
# (?<!\w) - na primeira parte, verifica se não há caracter apos do hifen -
```

```python
dados['descricao_local']
```

```python
#tokenização
dados['descricao_local'] = dados['descricao_local'].str.split()
dados['descricao_local']
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
```
